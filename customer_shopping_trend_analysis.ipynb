>## Q1. Which product categories sell most in each season?

>## Q2. Does discount/promo code usage affect purchase amount or review rating?

##### *.groupby('Had_Discount')* - Splits data into two groups: True (had discount/promo) and False (no discount/promo)
##### *.mean()* - Calculates average purchase amount for each group
##### *Boxplots* - Show the distribution (min, max, median, quartiles) for each group

```python
# Created a new combined column of Discount/PromoCode
df['Had_Discount'] = (df['Discount Applied'] == 'Yes') | (df['Promo Code Used'] == 'Yes')

# Compare averages
print(df.groupby('Had_Discount')['Purchase Amount (USD)'].mean())
print(df.groupby('Had_Discount')['Review Rating'].mean())
```

#### *subplots_adjust(wspace=0.6) :* sets the horizontal gap (whitespace) between multiple subplots to 60% of the average subplot width.

```python
plt.subplot(1,2,1)
sns.boxplot(data=df,x = 'Had_Discount',y = 'Purchase Amount (USD)')
plt.title("Discount affecting Purchase Amount (USD)")

plt.subplots_adjust(wspace=0.6)

plt.subplot(1,2,2)
sns.boxplot(data=df,x = 'Had_Discount',y = 'Review Rating')
plt.title("Discount affecting Review Rating")
```

>## Q3. How does Age and Season affect Purchase Amount?


```python
print(df["Age"].min())
print(df["Age"].max())
```

*.cut() :* bins continuous numerical values into discrete intervals or categories (converts a list of numbers into labeled groups or ranges.)

*Heatmap :* Colors show high vs low spending (darker = spends more)

*bins :* the cut points or edges that define where one group ends and another begins (bucket like).

```python
# Create age groups
df['Age_Groups']=pd.cut(df['Age'], bins = [18,30,40,50,60,71], labels = ['18-29','30-39','40-49','50-59','60-70'])

# Create pivot table for heatmap
heatmap_data = df.pivot_table(values='Purchase Amount (USD)', index='Age_Groups',  columns='Season', aggfunc='mean').round(0)
heatmap_data
```

```python
sns.heatmap(heatmap_data,annot= True)
```

>## Q4. What's the relationship between payment method and purchase amount?


```python
df.groupby('Payment Method')['Purchase Amount (USD)'].agg(['max','min','mean','median','sum','std','var','count']).round(2).sort_values('mean',ascending=False)
```

*plt.figure(figsize=(10,6))* --→ Creates a canvas of size 10x6 inches

*ordering* --→ Figures out which payment method has the highest average purchase amount, then sorts them from highest to lowest

*.index* --→ Extract just the payment method names (like ['Credit Card', 'PayPal', 'Cash']) without the average values

*sns.violinplot(... order=ordering)* --→ Draws violin shapes showing the full distribution (density) of purchase amounts for each payment method, arranged by highest spenders first

*sns.stripplot(... hue='Gender')* --→ Adds individual dots for each purchase, colored by gender, so you can see actual data points inside/on top of the violins

```python
plt.figure(figsize=(10,6))
ordering = df.groupby('Payment Method')['Purchase Amount (USD)'].mean().sort_values(ascending=False).index

sns.violinplot(data=df, x='Payment Method', y='Purchase Amount (USD)', order=ordering)
sns.stripplot(data=df, x='Payment Method', y='Purchase Amount (USD)', order=ordering, hue='Gender')
```

>## Q5. How does purchase frequency affect average spending?




*freq_order* = converting into desired order (Weekly → Bi-Weekly → Fortnightly...)

*reindex(freq_order)* = rearranges the grouped data to follow that exact order (Fixes the order of data to match a custom order you define.)

```python
# Defining correct order of frequency of purchase
freq_order = ['Weekly', 'Bi-Weekly', 'Fortnightly', 'Monthly', 'Every 3 Months', 'Quarterly', 'Annually']

# Group and reorder
freq_spending = df.groupby('Frequency of Purchases')['Purchase Amount (USD)'].mean().reindex(freq_order)
print(freq_spending)

plt.figure(figsize=(7,5))
plt.plot(freq_order, freq_spending.values, marker='o', linewidth=2, markersize=8, color='blue')
plt.title('Purchase Frequency vs Average Spending', fontsize=14)
plt.xlabel('Frequency of Purchases', fontsize=11)
plt.ylabel('Average Purchase Amount (USD $)', fontsize=11)
plt.xticks(rotation=25)
```

>## Q6. Which season has the highest sales and average order value?

*seasonal['sum']* --→ A Series with seasons as index and total sales as values

*.idxmax()* --→ Finds which season has the highest total sales, then returns that season name (returns the index/label of the maximum value in a Series.)

```python
seasonal=df.groupby('Season')['Purchase Amount (USD)'].agg(['mean','sum','count']).round(2)
print(seasonal)
# mean - Average order value per season
# sum - Total sales per season
# count - Number of order per season
  
print("\n")
  
# Find the winners
highest_sales_season = seasonal['sum'].idxmax()
highest_avg_season = seasonal['mean'].idxmax()

print(f"Highest Total Sales: {highest_sales_season}")
print(f"Highest Average Order: {highest_avg_season}")

bargraph = plt.barh(seasonal.index, seasonal['sum'],color='teal', edgecolor='black')
```

>#### Summary of graphs used --->
*BAR CHART* → Best for comparing categories → Shows exact values, easy to read differences

*BOXPLOT* → Best for distribution + outliers → Shows median, quartiles, min/max, and outliers

*HEATMAP* → Best for two-category patterns → Color intensity = value strength (e.g., average spending)

*VIOLIN PLOT* → Best for distribution shape → Shows where data is concentrated (density) + boxplot inside

*STRIP PLOT* → Best for showing individual data points → Helps see actual distribution, clustering, and overlaps (often layered with Violin Plot)

*LINE CHART* → Best for trends over order → Shows direction of change (e.g., frequency → spending)

*HORIZONTAL BAR* → Best for long category names → Easier to read when names are long

># ***STEP 4 : CREATING A DASHBOARD***

*plt.gca()* = Get Current Axes (the active plotting area)

*ax=plt.gca()* = Tell the plot to use the existing plotting area instead of creating a new one (It says, "Use the plot area I already set up")


```python
fig = plt.figure(figsize=(20, 12))
fig.suptitle('SHOPPING TRENDS DASHBOARD', fontsize=22, fontweight='bold')


plt.subplot(2,3,1)   #Q1
season_cat = df.groupby(["Season","Category"]).size().unstack()  
season_cat.plot(kind="bar", ax=plt.gca())  # ← Added ax=plt.gca()
plt.xticks(rotation=25)
plt.yticks(rotation=25)
plt.xlabel('Season'.upper(), fontsize=13)
plt.ylabel('Sales Count'.upper(), fontsize=13)
plt.title("Q1. Product Categories Sold by Season".upper(), fontsize=18)
              
plt.subplot(2,3,2)   #Q2
sns.boxplot(data=df,x = 'Had_Discount',y = 'Purchase Amount (USD)')
plt.title("Q2. Discount Effect on Purchase Amount".upper(), fontsize=18)
plt.xlabel('Had Discount/Promo code'.upper(), fontsize=13)
plt.ylabel('Purchase Amount (USD)'.upper(), fontsize=13)

plt.subplot(2,3,3)   #Q3
df['Age_Groups']=pd.cut(df['Age'], bins = [18,30,40,50,60,71], labels = ['18-29','30-39','40-49','50-59','60-70'])
# Create pivot table for heatmap
heatmap_data = df.pivot_table(values='Purchase Amount (USD)', index='Age_Groups',  columns='Season', aggfunc='mean').round(0)
sns.heatmap(heatmap_data,annot=True)
plt.xlabel('SEASON', fontsize=13)
plt.ylabel('Age Groups'.upper(), fontsize=13)
plt.title("Q3. Age Group vs Seasonal Spending".upper(), fontsize=18)

plt.subplot(2,3,4)    #Q4
ordering = df.groupby('Payment Method')['Purchase Amount (USD)'].mean().sort_values(ascending=False).index
sns.violinplot(data=df, x='Payment Method', y='Purchase Amount (USD)', order=ordering)
sns.stripplot(data=df, x='Payment Method', y='Purchase Amount (USD)', order=ordering, hue='Gender')
plt.xlabel('Payment Method'.upper(), fontsize=13)
plt.ylabel('Purchase Amount (USD)'.upper(), fontsize=13)
plt.title("Q4. Payment Method vs Purchase Amount".upper(), fontsize=18)

plt.subplot(2,3,5)    #Q5
# Defining correct order of frequency of purchase
freq_order = ['Weekly', 'Bi-Weekly', 'Fortnightly', 'Monthly', 'Every 3 Months', 'Quarterly', 'Annually']
# Group and reorder
freq_spending = df.groupby('Frequency of Purchases')['Purchase Amount (USD)'].mean().reindex(freq_order)
plt.plot(freq_order, freq_spending.values, marker='o', linewidth=2, markersize=8, color='blue')
plt.xlabel('Frequency of Purchases'.upper(), fontsize=13)
plt.ylabel('Average Purchase Amount (USD)'.upper(), fontsize=13)
plt.xticks(rotation=25)
plt.title("Q5. Purchase Frequency vs Average Spending".upper(), fontsize=18)

plt.subplot(2,3,6)    #Q6       
seasonal=df.groupby('Season')['Purchase Amount (USD)'].agg(['mean','sum','count']).round(2)
print("\n")
bargraph = plt.barh(seasonal.index, seasonal['sum'],color='teal', edgecolor='black')
plt.ylabel('SEASON', fontsize=13)
plt.xlabel('TOTAL SALES BY SEASON', fontsize=13)
plt.title("Q6. SEASON WITH HIGHEST SALES", fontsize=18)

plt.tight_layout()
plt.subplots_adjust(hspace=0.3, wspace=0.3, top=0.92)
plt.show()
```
